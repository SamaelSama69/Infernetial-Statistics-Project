### 1.1 What is the probability that a randomly chosen player would suffer an injury?

```python
# Calculate the total number of injured players
total_injured = df_problem1["Players Injured"].sum()

# Calculate the total number of players (both injured and not injured)
total_players = df_problem1[["Players Injured", "Players Not Injured"]].sum().sum()

# Calculate the probability that a randomly chosen player would suffer an injury
prob_injury = total_injured / total_players

# Print the resulting probability in a user-friendly format
print("The probability that a randomly chosen player would suffer an injury is:", prob_injury)
```

### 1.2 What is the probability that a player is a forward or a winger?

```python
# Calculate the total number of players in the "Forward" position
frds = (df_problem1.loc[df_problem1["Position"] == "Forward", "Players Injured"].values[0] +
        df_problem1.loc[df_problem1["Position"] == "Forward", "Players Not Injured"].values[0])

# Calculate the total number of players in the "Winger" position
wngs = (df_problem1.loc[df_problem1["Position"] == "Winger", "Players Injured"].values[0] +
        df_problem1.loc[df_problem1["Position"] == "Winger", "Players Not Injured"].values[0])

# Calculate the total number of players (both injured and not injured)
total_players = df_problem1[["Players Injured", "Players Not Injured"]].sum().sum()

# Calculate the probability that a player is either a "Forward" or a "Winger"
prob_frds_or_wngs = (frds + wngs) / total_players

# Print the result
print("The probability that a player is a forward or a winger is:", prob_frds_or_wngs)
```

### 1.3 What is the probability that a randomly chosen player plays in a striker position and has a foot injury?

```python
# Extract the number of injured players in the "Striker" position
stkr_injured = df_problem1.loc[df_problem1["Position"] == "Striker", "Players Injured"].values[0]

# Calculate the total number of players (both injured and not injured)
total_players = df_problem1[["Players Injured", "Players Not Injured"]].sum().sum()

# Calculate the probability that a randomly chosen player is a striker and has a foot injury
prob_stkr_foot_injury = stkr_injured / total_players

# Print the calculated probability
print("Probability that a randomly chosen player plays in a striker position and has a foot injury:", prob_stkr_foot_injury)
```

### 1.4 What is the probability that a randomly chosen injured player is a striker?

```python
# Calculate the total number of injured players
total_injured_players = df_problem1["Players Injured"].sum()

# Calculate the probability that a randomly chosen injured player is a striker
prob_injured_stkr = stkr_injured / total_injured_players

# Print the calculated probability
print("Probability that a randomly chosen injured player is a striker:", prob_injured_stkr)
```

## Problem 2
### The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain

### 2.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
# Define the mean (mu) and standard deviation (sigma) of the normal distribution
mu = 5  # Mean breaking strength in kg/cm²
sigma = 1.5  # Standard deviation of breaking strength in kg/cm²

# Specify the breaking strength threshold
breaking_strength1 = 3.17  # Breaking strength in kg/cm²

# Calculate the cumulative probability for breaking strength < breaking_strength1
prop_1 = stats.norm.cdf(breaking_strength1, mu, sigma)

# Display the proportion in a formatted output
print(f"Proportion with breaking strength < {breaking_strength1} kg/cm²: {prop_1:.4f}")
```

```python
# Define the range of x values for the normal distribution
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# Calculate the Probability Density Function (PDF) for the normal distribution
pdf = stats.norm.pdf(x, mu, sigma)

# Plot the PDF of the normal distribution
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# Highlight the breaking strength threshold with a vertical red line
plt.axvline(x=breaking_strength1, color="r", label=f'Breaking Strength = {breaking_strength1} kg/cm²')

# Define the range for the filled region under the curve
x_fill = np.linspace(x.min(), breaking_strength1, 1000)

# Fill the area under the curve for P(X < breaking_strength1)
plt.fill_between(
    x_fill,
    stats.norm.pdf(x_fill, mu, sigma),
    color="red",
    alpha=0.3,
    label=f'P(X < {breaking_strength1}) = {prop_1:.4f}'
)

# Add labels and a title to the plot
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Probability Density")
plt.title("Normal Distribution of Gunny Bag Breaking Strength")

# Add a legend to explain the plot elements
plt.legend()

# Display the plot
plt.show()
```

### 2.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
# Define the mean (mu) and standard deviation (sigma) of the normal distribution
mu = 5  # Mean breaking strength in kg/cm²
sigma = 1.5  # Standard deviation of breaking strength in kg/cm²

# Specify the breaking strength threshold
breaking_strength2 = 3.6  # Breaking strength in kg/cm²

# Calculate the proportion of gunny bags with breaking strength >= breaking_strength2
prop_2 = 1 - stats.norm.cdf(breaking_strength2, mu, sigma)

# Print the proportion in a user-friendly format
print(f"Proportion with breaking strength >= {breaking_strength2} kg/cm²: {prop_2:.4f}")
```

```python
# Create a figure with defined size for better visualization
plt.figure(figsize=(10, 6))

# Plot the Probability Density Function (PDF) of the normal distribution
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# Highlight the breaking strength threshold with a vertical red line
plt.axvline(x=breaking_strength2, color="r", label=f'Breaking Strength = {breaking_strength2} kg/cm²')

# Define the range for the filled area under the curve
x_fill = np.linspace(breaking_strength2, x.max(), 1000)

# Fill the area under the curve from breaking_strength2 to the end of the range
plt.fill_between(
    x_fill,
    stats.norm.pdf(x_fill, mu, sigma),
    color="red",
    alpha=0.3,
    label=f'P(X >= {breaking_strength2}) = {prop_2:.4f}'
)

# Add labels for the x-axis and y-axis
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Probability Density")

# Add a title to the plot
plt.title("Normal Distribution of Gunny Bag Breaking Strength")

# Add a legend to explain the elements of the plot
plt.legend()

# Display the plot
plt.show()
```

### 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
# Define the bounds for breaking strength
l_bound1 = 5       # Lower bound for breaking strength in kg/cm²
u_bound1 = 5.5     # Upper bound for breaking strength in kg/cm²

# Calculate the proportion of gunny bags with breaking strength between the bounds
prop_3 = stats.norm.cdf(u_bound1, mu, sigma) - stats.norm.cdf(l_bound1, mu, sigma)

# Print the calculated proportion in a user-friendly format
print(f"Proportion with breaking strength between {l_bound1} and {u_bound1} kg/cm²: {prop_3:.4f}")
```

```python
# Set up the figure for the plot with specified dimensions
plt.figure(figsize=(10, 6))

# Plot the Probability Density Function (PDF) of the normal distribution
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# Add a vertical dashed red line for the lower bound of the breaking strength
plt.axvline(x=l_bound1, color="r", linestyle='--', label=f'l_bound1 = {l_bound1} kg/cm²')

# Add a vertical dashed green line for the upper bound of the breaking strength
plt.axvline(x=u_bound1, color="g", linestyle='--', label=f'u_bound1 = {u_bound1} kg/cm²')

# Define the range for the filled area between the lower and upper bounds
x_fill = np.linspace(l_bound1, u_bound1, 1000)

# Fill the area under the curve between the bounds with an orange color and semi-transparency
plt.fill_between(
    x_fill,
    stats.norm.pdf(x_fill, mu, sigma),
    color="orange",
    alpha=0.5
)

# Add labels for the axes
plt.xlabel("Breaking Strength (kg/cm²)")
plt.ylabel("Probability Density")

# Add a title to the plot
plt.title("Normal Distribution of Gunny Bag Breaking Strength")

# Add a legend to the plot to describe the elements
plt.legend()

# Display the plot
plt.show()
```

### 2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
# Define the lower and upper bounds for breaking strength
l_bound2 = 3      # Lower bound for breaking strength in kg/cm²
u_bound2 = 7.5    # Upper bound for breaking strength in kg/cm²

# Calculate the proportion of gunny bags with breaking strength NOT between the bounds
# This is calculated as the sum of probabilities outside the interval (l_bound2, u_bound2)
prop_4 = stats.norm.cdf(l_bound2, mu, sigma) + (1 - stats.norm.cdf(u_bound2, mu, sigma))

# Print the calculated proportion in a user-friendly format
print(f"Proportion with breaking strength NOT between {l_bound2} and {u_bound2} kg/cm²: {prop_4:.4f}")
```

```python
# Create a figure with specified dimensions for better visualization
plt.figure(figsize=(10, 6))

# Plot the Probability Density Function (PDF) of the normal distribution
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# Add a vertical dashed red line to indicate the lower bound of breaking strength
plt.axvline(x=l_bound2, color="r", linestyle='--', label=f'l_bound2 = {l_bound2} kg/cm²')

# Add a vertical dashed green line to indicate the upper bound of breaking strength
plt.axvline(x=u_bound2, color="g", linestyle='--', label=f'u_bound2 = {u_bound2} kg/cm²')

# Fill the area under the curve outside the bounds (x <= l_bound2 or x >= u_bound2)
plt.fill_between(
    x,
    0,
    pdf,
    where=(x <= l_bound2) | (x >= u_bound2),
    color='orange',
    alpha=0.3,
    label=f'Area NOT between {l_bound2} and {u_bound2} kg/cm²'
)

# Add a title to the plot
plt.title('Normal Distribution of Breaking Strength (P(X NOT between 3 and 7.5 kg/cm²))')

# Add labels for the x-axis and y-axis
plt.xlabel('Breaking Strength (kg/cm²)')
plt.ylabel('Density')

# Add a legend to describe the elements of the plot
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

### 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?
##### "- State the null and alternate hypotheses - Conduct the hypothesis test and compute the p-value - Write down conclusions from the test results Note: Consider the level of significance as 5%."

#### Null Hypothesis (H 0​): The mean Brinell hardness of unpolished stones is at least 150.
#### Alternative Hypothesis (H 1​): The mean Brinell hardness of unpolished stones is less than 150.

```python
sample_mean = 134.33
sample_std = 30.62
n = 75
population_mean = 150
alpha = 0.05
```

```python
# Perform a one-sample t-test on the "Unpolished" column
# Null Hypothesis: The mean Brinell hardness of unpolished stones is at least 150.
# Alternative Hypothesis: The mean Brinell hardness of unpolished stones is less than 150.

# Extract the "Unpolished" column
unpolished_hardness = df_problem3['Unpolished']

# Conduct a one-sample t-test to compare the sample mean against the hypothesized value of 150
t_stat, p_value = stats.ttest_1samp(unpolished_hardness, 150)

# Calculate the one-tailed p-value by dividing the two-tailed p-value by 2
p_value_one_tailed = p_value / 2

# Print the t-statistic and one-tailed p-value in a formatted output
print(f"One-Sample T-Test for Unpolished Stones: t-statistic = {t_stat:.4f}, p-value (one-tailed) = {p_value_one_tailed:.4f}")

# Evaluate the test results based on the significance level (alpha = 0.05)
if p_value_one_tailed < 0.05 and t_stat < 0:
    # Conclusion if p-value is significant and t-statistic indicates a mean less than 150
    print("Zingaro is justified in thinking the unpolished stones may not be suitable for printing.")
else:
    # Conclusion if the null hypothesis cannot be rejected
    print("Zingaro is not justified in thinking the unpolished stones may not be suitable for printing.")
```

### 3.2 Is the mean hardness of the polished and unpolished stones the same?

#### Null Hypothesis : The mean hardness of polished stones is equal to the mean hardness of unpolished stones

#### Alternative Hypothesis : The mean hardness of polished stones is not equal to the mean hardness of unpolished stones.

```python
# Extract the 'Unpolished' and 'Treated and Polished' columns from the dataset
unpolished_hardness = df_problem3['Unpolished']
polished_hardness = df_problem3['Treated and Polished']

# Perform a two-sample t-test
# Null Hypothesis (H0): The mean hardness of polished stones is equal to the mean hardness of unpolished stones.
# Alternative Hypothesis (H1): The mean hardness of polished stones is not equal to the mean hardness of unpolished stones.
t_stat, p_value = stats.ttest_ind(unpolished_hardness, polished_hardness)

# Print the t-statistic and p-value in a well-formatted output
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Compare the p-value against the significance level (alpha = 0.05)
if p_value < alpha:
    # Conclusion if p-value is less than alpha (reject null hypothesis)
    print("Reject the null hypothesis: The mean hardness of polished and unpolished stones is significantly different.")
else:
    # Conclusion if p-value is greater than or equal to alpha (fail to reject null hypothesis)
    print("Fail to reject the null hypothesis: There is not enough evidence to say the mean hardness of polished and unpolished stones is different.")
```

```python
# Set up the figure for the boxplot with specified dimensions
plt.figure(figsize=(10, 6))

# Create a boxplot to compare the hardness of unpolished and treated/polished stones
sns.boxplot(
    data=[unpolished_hardness, polished_hardness],  # Data for the boxplot
)

# Add a title to the boxplot
plt.title('Comparison of Hardness: Unpolished vs Treated and Polished Stones')

# Label the x-axis to represent the stone types
plt.xlabel('Stone Type')

# Label the y-axis to represent hardness values
plt.ylabel('Hardness')

# Customize the x-axis ticks to label each group
plt.xticks(
    ticks=[0, 1],                  # Tick positions
    labels=['Unpolished', 'Treated and Polished']  # Labels for the stone types
)

# Display the plot
plt.show()
```
